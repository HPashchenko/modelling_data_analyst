```python
#групуємо дані та визначаємо кількість товарів по містам та департаментам
df.groupby(by=['Branch', 'City'])['Quantity'].sum()
```

```python
# зведена таблиця кількості категоріі товарів по містам
quantity_by_category = pd.pivot_table(df, index = 'City', columns = 'Product line', values = 'Quantity', aggfunc='sum')
quantity_by_category
```

```python
from timeit import default_timer as timer
# Ковертую  колонку Date та виділяю лише місяць
temp = pd.DatetimeIndex(df['Date'])
```

```python
df['Month'] = temp.month
```

```python
df['Month'].replace(1, 'January', inplace=True)
df['Month'].replace(2, 'February', inplace=True)
df['Month'].replace(3, 'March', inplace=True)
df['Month'].replace(4, 'April', inplace=True)
df['Month'].replace(5, 'May', inplace=True)
df['Month'].replace(6, 'June', inplace=True)
df['Month'].replace(7, 'July', inplace=True)
df['Month'].replace(8, 'August', inplace=True)
df['Month'].replace(9, 'September', inplace=True)
df['Month'].replace(10, 'October', inplace=True)
df['Month'].replace(11, 'November', inplace=True)
df['Month'].replace(12, 'December', inplace=True)
```

```python
df['Month'].nunique()
```

```python
#зведена таблиця кількості товарів, проданих потягом трьох місяців, згідно міст
buy_by_month_pivot = pd.pivot_table(df, index = 'Month', columns = 'City', values = 'Quantity', aggfunc='sum', sort=False)
buy_by_month_pivot
```

```python
buy_per_month = df.groupby(by=['Month', 'City', 'Product line'])['Quantity'].sum()
buy_per_month
```
